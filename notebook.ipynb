# Aircraft analysis


```python
import pandas as pd

df = pd.read_csv("/content/sample_data/aircrafts.csv", skiprows=0)
df.head()
import matplotlib.pyplot as plt
```

# Allocation of Seats on Aircrafts



```python
df.hist(column='seats', bins=20, grid=False, figsize=(10,6), color='#86bf91', zorder=2, rwidth=0.9)

plt.title('Distribuição de Assentos nas Aeronaves')
plt.xlabel('Assentos')
plt.ylabel('Número de Aeronaves')
plt.show()
```

# Distribution of Years of Manufacture of Aircraft

```python
df.hist(column='year', bins=20, grid=False, figsize=(10,6), color='#86bf91', zorder=2, rwidth=0.9)

plt.title('Distribuição de Anos de Fabricação das Aeronaves')
plt.xlabel('Ano')
plt.ylabel('Número de Aeronaves')
plt.show()
```

# Aircraft Type Frequency

```python
freq = df['type'].value_counts()

freq.plot(kind='bar', color='#86bf91', figsize=(10,6))

plt.title('Frequência de Tipos de Aeronaves')
plt.xlabel('Tipo')
plt.ylabel('Frequência')
plt.show()
```

# Relationship between the Number of Engines and the Number of Seats

```python
df.plot(kind='scatter', x='engines', y='seats', color='#86bf91', figsize=(10,6))

plt.title('Relação entre o Número de Motores e o Número de Assentos')
plt.xlabel('Número de Motores')
plt.ylabel('Número de Assentos')
plt.show()
```

# Relationship between Models and Seats

```python
model_seats = df.groupby('model')['seats'].mean().reset_index()

# Ordena os dados pela média de assentos em ordem decrescente
model_seats = model_seats.sort_values(by='seats', ascending=False)

# Cria um gráfico de barras para visualizar a relação entre modelos e assentos
fig, ax = plt.subplots(figsize=(20, 10))

ax.bar(model_seats['model'], model_seats['seats'], color='#86bf91')
ax.set_title('Relação entre Modelos e Assentos')
ax.set_xlabel('Modelo')
ax.set_ylabel('Média de Assentos')
ax.tick_params(axis='x', rotation=90)

plt.show()
```
